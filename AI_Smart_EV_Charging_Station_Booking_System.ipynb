```python
# ==========================================================
# AI Smart EV Charging Station Booking System
# Module 1: Database Setup
# ==========================================================

# Import SQLite library
import sqlite3

# Create or connect to the SQLite database
conn = sqlite3.connect("ev_charging_system.db")

# Create a cursor object
cursor = conn.cursor()

# Display success message
print("========================================")
print("SQLite Database Initialized Successfully!")
print("Database Name : ev_charging_system.db")
print("Connection Status : Connected")
print("========================================")
```

```python
# ==========================================================
# AI Smart EV Charging Station Booking System
# Module 1 - Database Setup
# Step 2 - Create Database Tables
# ==========================================================

# -------------------------
# Create Users Table
# -------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Users (
    user_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    phone TEXT,
    vehicle_model TEXT,
    battery_capacity INTEGER
)
""")

# -------------------------
# Create Charging Stations Table
# -------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS ChargingStations (
    station_id INTEGER PRIMARY KEY AUTOINCREMENT,
    station_name TEXT NOT NULL,
    location TEXT NOT NULL,
    available_chargers INTEGER,
    charging_speed TEXT,
    price_per_kwh REAL
)
""")

# -------------------------
# Create Bookings Table
# -------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Bookings (
    booking_id INTEGER PRIMARY KEY AUTOINCREMENT,
    user_id INTEGER,
    station_id INTEGER,
    booking_date TEXT,
    booking_time TEXT,
    status TEXT,
    FOREIGN KEY (user_id) REFERENCES Users(user_id),
    FOREIGN KEY (station_id) REFERENCES ChargingStations(station_id)
)
""")

# -------------------------
# Create Payments Table
# -------------------------
cursor.execute("""
CREATE TABLE IF NOT EXISTS Payments (
    payment_id INTEGER PRIMARY KEY AUTOINCREMENT,
    booking_id INTEGER,
    amount REAL,
    payment_status TEXT,
    FOREIGN KEY (booking_id) REFERENCES Bookings(booking_id)
)
""")

# Save changes
conn.commit()

# Display success message
print("========================================")
print("All database tables created successfully!")
print("Tables Created:")
print("1. Users")
print("2. ChargingStations")
print("3. Bookings")
print("4. Payments")
print("========================================")
```

```python
# ==========================================================
# AI Smart EV Charging Station Booking System
# Core Data Function - Search Charging Stations
# ==========================================================

import pandas as pd

# ----------------------------------------------------------
# Function to Search Charging Stations by Location
# ----------------------------------------------------------
def search_charging_station():

    # Get location from user
    location = input("Enter Location to Search: ")

    # Execute SQL query (case-insensitive search)
    cursor.execute("""
        SELECT
            station_id,
            station_name,
            location,
            available_chargers,
            charging_speed,
            price_per_kwh
        FROM ChargingStations
        WHERE LOWER(location) = LOWER(?)
    """, (location,))

    records = cursor.fetchall()

    # Display results
    if records:

        columns = [
            "Station ID",
            "Station Name",
            "Location",
            "Available Chargers",
            "Charging Speed",
            "Price per kWh"
        ]

        df = pd.DataFrame(records, columns=columns)

        print("\n==========================================")
        print("     SEARCH RESULTS")
        print("==========================================")

        print(df.to_string(index=False))

    else:

        print("\nNo charging station found for the entered location.")

# ----------------------------------------------------------
# Test the Search Function
# ----------------------------------------------------------
search_charging_station()
```

```python
# ==========================================================
# AI Smart EV Charging Station Booking System
# Core Data Function - Filter Charging Stations
# ==========================================================

import pandas as pd

# ----------------------------------------------------------
# Function to Filter Charging Stations by Minimum Speed
# ----------------------------------------------------------
def filter_charging_stations():

    # Get minimum charging speed from the user
    min_speed = int(input("Enter Minimum Charging Speed (kW): "))

    # Execute SQL query
    cursor.execute("""
        SELECT
            station_id,
            station_name,
            location,
            available_chargers,
            charging_speed,
            price_per_kwh
        FROM ChargingStations
    """)

    records = cursor.fetchall()

    filtered_records = []

    # Filter stations based on charging speed
    for record in records:
        speed = int(record[4].replace(" kW", ""))

        if speed >= min_speed:
            filtered_records.append(record)

    # Display results
    if filtered_records:

        columns = [
            "Station ID",
            "Station Name",
            "Location",
            "Available Chargers",
            "Charging Speed",
            "Price per kWh"
        ]

        df = pd.DataFrame(filtered_records, columns=columns)

        print("\n==========================================")
        print("     FILTERED CHARGING STATIONS")
        print("==========================================")

        print(df.to_string(index=False))

    else:

        print("\nNo charging stations match the selected charging speed.")

# ----------------------------------------------------------
# Test the Filter Function
# ----------------------------------------------------------
filter_charging_stations()
```

```python
# ==========================================================
# AI Smart EV Charging Station Booking System
# Main Business Function - Book Charging Slot
# ==========================================================

# ----------------------------------------------------------
# Function to Book a Charging Slot
# ----------------------------------------------------------

def book_charging_slot():

    # -----------------------------
    # Get Booking Details
    # -----------------------------
    user_id = int(input("Enter User ID: "))
    station_id = int(input("Enter Charging Station ID: "))
    booking_date = input("Enter Booking Date (YYYY-MM-DD): ")
    booking_time = input("Enter Booking Time (HH:MM AM/PM): ")

    # -----------------------------
    # Validate User
    # -----------------------------
    cursor.execute(
        "SELECT * FROM Users WHERE user_id = ?",
        (user_id,)
    )

    if cursor.fetchone() is None:
        print("\nInvalid User ID!")
        return

    # -----------------------------
    # Validate Charging Station
    # -----------------------------
    cursor.execute(
        "SELECT available_chargers FROM ChargingStations WHERE station_id = ?",
        (station_id,)
    )

    station = cursor.fetchone()

    if station is None:
        print("\nInvalid Charging Station ID!")
        return

    available_chargers = station[0]

    # -----------------------------
    # Check Charger Availability
    # -----------------------------
    if available_chargers <= 0:
        print("\nNo chargers available at this station.")
        return

    # -----------------------------
    # Insert Booking
    # -----------------------------
    cursor.execute("""
        INSERT INTO Bookings
        (user_id, station_id, booking_date, booking_time, status)
        VALUES (?, ?, ?, ?, ?)
    """, (
        user_id,
        station_id,
        booking_date,
        booking_time,
        "Confirmed"
    ))

    booking_id = cursor.lastrowid

    # -----------------------------
    # Update Available Chargers
    # -----------------------------
    cursor.execute("""
        UPDATE ChargingStations
        SET available_chargers = available_chargers - 1
        WHERE station_id = ?
    """, (station_id,))

    # Save Changes
    conn.commit()

    # -----------------------------
    # Booking Confirmation
    # -----------------------------
    print("\n==============================================")
    print("        BOOKING CONFIRMED")
    print("==============================================")
    print(f"Booking ID          : {booking_id}")
    print(f"User ID             : {user_id}")
    print(f"Charging Station ID : {station_id}")
    print(f"Booking Date        : {booking_date}")
    print(f"Booking Time        : {booking_time}")
    print("Booking Status      : Confirmed")
    print("==============================================")

# ----------------------------------------------------------
# Execute Function
# ----------------------------------------------------------

book_charging_slot()
```

```python
# ==========================================================
# AI Smart EV Charging Station Booking System
# Phase 7 - Maintenance & Prediction
# Step 1 - Generate 30 Days of Server Metrics
# ==========================================================

import random
import pandas as pd

# ----------------------------------------------------------
# Generate 30 Days of Server Performance Metrics
# ----------------------------------------------------------

server_metrics = []

for day in range(1, 31):

    daily_users = random.randint(180, 450)
    response_time = random.randint(120, 280)      # milliseconds
    database_load = random.randint(35, 90)        # percentage
    error_rate = round(random.uniform(0.2, 4.5), 2)  # percentage

    server_metrics.append([
        day,
        daily_users,
        response_time,
        database_load,
        error_rate
    ])

# ----------------------------------------------------------
# Create DataFrame
# ----------------------------------------------------------

columns = [
    "Day",
    "Daily Active Users",
    "API Response Time (ms)",
    "Database Load (%)",
    "Error Rate (%)"
]

metrics_df = pd.DataFrame(server_metrics, columns=columns)

# ----------------------------------------------------------
# Display Server Metrics
# ----------------------------------------------------------

print("=" * 80)
print("      AI SMART EV CHARGING STATION BOOKING SYSTEM")
print("        PHASE 7 - SERVER PERFORMANCE METRICS")
print("=" * 80)

print("\n30 DAYS OF SERVER PERFORMANCE DATA\n")

print(metrics_df.to_string(index=False))

print("\n" + "=" * 80)
print("30 days of server metrics generated successfully.")
print("The dataset has been stored for prediction and maintenance analysis.")
print("=" * 80)
```

```python
# ==========================================================
# AI Smart EV Charging Station Booking System
# Phase 7 - Maintenance & Prediction
# Step 3 - Predict Critical System Thresholds
# ==========================================================

import pandas as pd

# ----------------------------------------------------------
# Calculate Average Metrics
# ----------------------------------------------------------

avg_response = metrics_df["API Response Time (ms)"].mean()
avg_database = metrics_df["Database Load (%)"].mean()
avg_error = metrics_df["Error Rate (%)"].mean()

# ----------------------------------------------------------
# Define Critical Thresholds
# ----------------------------------------------------------

thresholds = {
    "API Response Time (ms)": 300,
    "Database Load (%)": 95,
    "Error Rate (%)": 5
}

# ----------------------------------------------------------
# Estimate Remaining Days
# ----------------------------------------------------------

prediction_data = []

metrics = [
    ("API Response Time (ms)", avg_response),
    ("Database Load (%)", avg_database),
    ("Error Rate (%)", avg_error)
]

for metric_name, current_value in metrics:

    threshold = thresholds[metric_name]

    difference = threshold - current_value

    # Simple prediction model
    estimated_days = max(1, int(difference / 2))

    prediction_data.append([
        metric_name,
        round(current_value, 2),
        threshold,
        estimated_days
    ])

# ----------------------------------------------------------
# Create Prediction Table
# ----------------------------------------------------------

prediction_df = pd.DataFrame(
    prediction_data,
    columns=[
        "Metric",
        "Current Average",
        "Critical Threshold",
        "Estimated Days Remaining"
    ]
)

# ----------------------------------------------------------
# Display Prediction Report
# ----------------------------------------------------------

print("=" * 90)
print("        AI SMART EV CHARGING STATION BOOKING SYSTEM")
print("      PREDICTION OF CRITICAL SYSTEM THRESHOLDS")
print("=" * 90)

print("\nAI PREDICTION REPORT\n")

print(prediction_df.to_string(index=False))

print("\n" + "=" * 90)
print("AI Prediction Completed Successfully!")
print("=" * 90)
```
